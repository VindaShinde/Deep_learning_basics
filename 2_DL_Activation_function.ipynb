```python
import numpy as np
from numpy import array, random, dot
from random import choice
```

```python
#initializing data set
dataset = [
    (array ([0,0,1]), 0), #array (first 2 input values, 3rd value is bias, last is the output value for superwised learning)
    (array ([0,1,1]), 1),
    (array ([1,0,1]), 1),
    (array ([1,1,1]), 1)
    
]
print(dataset)
```

```python
#initializing Random number for WEIGHTS
weights = random.rand(3);weights
```

```python
#initializing additional variables
r = 0.2 # learning rate
n = 100 # number of iterations
```

# Activation Function

LINEAR activation function (input = output)

```python
# Training and Evaluating using LINEAR avtivation function

activationFn = lambda x:x

for j in range(n):
  x, expected = choice(dataset) #random input set, includes repeated numbers
  result = dot(weights,x) #dot product 
  err = expected - activationFn(result)
  weights += r * err * x

for x, _ in dataset:
  result = dot(x,weights)
  print("Input:{} ResultBAFn:{} ResultAFn: {}".format(x[:2], round(result,3), activationFn(result)))
```

## STEP activation function (output 0 if x <0 and 1 is x> 0)

```python
# Training and Evaluating using STEP avtivation function
activationFn = lambda x: 0 if x < 0 else 1

for j in range(n):
  x, expected = choice(dataset) #random input set, includes repeated numbers
  result = dot(weights,x) #dot product 
  err = expected - activationFn(result)
  weights += r * err * x

for x, _ in dataset:
  result = dot(x,weights)
  print("Input:{} ResultBAFn:{} ResultAFn: {}".format(x[:2], round(result,3), activationFn(result)))
```

### SIGMOID activation function

![image.png](Sigmoid fuction.png)

```python
activationFn = lambda x: 1/(1+np.exp(-x))
err = []

for j in range(n):
  x, expected = choice(dataset) #random input set, includes repeated numbers
  result = dot(weights,x) #dot product 
  err = expected - activationFn(result)
  weights += r * err * x

for x, _ in dataset:
  result = dot(x,weights)
  print("Input:{} ResultBAFn:{} ResultAFn: {}".format(x[:2], round(result,3), activationFn(result)))
```

## RELU - Rectified Linear Unit Activation function
(Return 0 if x(input) is less than 0, returns x if x(input) is greater than 0
image.png

```python
activationFn = lambda x: 0 if x <0 else x

for j in range(n):
  x, expected = choice(dataset) #random input set, includes repeated numbers
  result = dot(weights,x) #dot product 
  err = expected - activationFn(result)
  weights += r * err * x

for x, _ in dataset:
  result = dot(x,weights)
  print("Input:{} ResultBAFn:{} ResultAFn: {}".format(x[:2], round(result,3), activationFn(result)))
```

## SOFTMAX activation function (Probability score)

image.png

```python
activationFn = lambda x: np.exp(-x)/ np.sum(np.exp(x), axis =0)
err = []
for j in range(n):
  x, expected = choice(dataset) #random input set, includes repeated numbers
  result = dot(weights,x) #dot product 
  err = expected - activationFn(result)
  weights += r * err * x

for x, _ in dataset:
  result = dot(x,weights)
  print("Input:{} ResultBAFn:{} ResultAFn: {}".format(x[:2], round(result,3), activationFn(result)))
```
